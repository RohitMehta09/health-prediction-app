```python
heart_encoded.fillna(heart_encoded.mean(), inplace=True)
diabetes_df.fillna(diabetes_df.mean(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

# For Heart Dataset
X_heart = heart_encoded.drop('HeartDisease', axis=1)
y_heart = heart_encoded['HeartDisease']

scaler_heart = StandardScaler()
X_heart_scaled = scaler_heart.fit_transform(X_heart)

# For Diabetes Dataset
X_diabetes = diabetes_df.drop('Outcome', axis=1)
y_diabetes = diabetes_df['Outcome']

scaler_diabetes = StandardScaler()
X_diabetes_scaled = scaler_diabetes.fit_transform(X_diabetes)
```

```python
from sklearn.model_selection import train_test_split

# Heart
Xh_train, Xh_test, yh_train, yh_test = train_test_split(X_heart_scaled, y_heart, test_size=0.2, random_state=42)

# Diabetes
Xd_train, Xd_test, yd_train, yd_test = train_test_split(X_diabetes_scaled, y_diabetes, test_size=0.2, random_state=42)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

# Train Models on Heart Dataset

```python
print("----- Heart Disease Model Training -----")

# Logistic Regression
lr_heart = LogisticRegression()
lr_heart.fit(Xh_train, yh_train)
pred_lr_heart = lr_heart.predict(Xh_test)

# Random Forest
rf_heart = RandomForestClassifier()
rf_heart.fit(Xh_train, yh_train)
pred_rf_heart = rf_heart.predict(Xh_test)

# SVM
svm_heart = SVC()
svm_heart.fit(Xh_train, yh_train)
pred_svm_heart = svm_heart.predict(Xh_test)

# Evaluate
models_heart = {'Logistic Regression': pred_lr_heart, 'Random Forest': pred_rf_heart, 'SVM': pred_svm_heart}

for name, preds in models_heart.items():
    print(f"\n{name} Results:")
    print("Accuracy:", accuracy_score(yh_test, preds))
    print("Confusion Matrix:\n", confusion_matrix(yh_test, preds))
    print("Classification Report:\n", classification_report(yh_test, preds))
```

# Train Models on Diabetes Dataset

```python
print("\n\n----- Diabetes Model Training -----")

# Logistic Regression
lr_diabetes = LogisticRegression()
lr_diabetes.fit(Xd_train, yd_train)
pred_lr_diabetes = lr_diabetes.predict(Xd_test)

# Random Forest
rf_diabetes = RandomForestClassifier()
rf_diabetes.fit(Xd_train, yd_train)
pred_rf_diabetes = rf_diabetes.predict(Xd_test)

# SVM
svm_diabetes = SVC()
svm_diabetes.fit(Xd_train, yd_train)
pred_svm_diabetes = svm_diabetes.predict(Xd_test)

# Evaluate
models_diabetes = {'Logistic Regression': pred_lr_diabetes, 'Random Forest': pred_rf_diabetes, 'SVM': pred_svm_diabetes}

for name, preds in models_diabetes.items():
    print(f"\n{name} Results:")
    print("Accuracy:", accuracy_score(yd_test, preds))
    print("Confusion Matrix:\n", confusion_matrix(yd_test, preds))
    print("Classification Report:\n", classification_report(yd_test, preds))
```

```python
def plot_conf_matrix(true, pred, title):
    cm = confusion_matrix(true, pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Heart Models
plot_conf_matrix(yh_test, pred_lr_heart, 'Heart - Logistic Regression')
plot_conf_matrix(yh_test, pred_rf_heart, 'Heart - Random Forest')
plot_conf_matrix(yh_test, pred_svm_heart, 'Heart - SVM')

# Diabetes Models
plot_conf_matrix(yd_test, pred_lr_diabetes, 'Diabetes - Logistic Regression')
plot_conf_matrix(yd_test, pred_rf_diabetes, 'Diabetes - Random Forest')
plot_conf_matrix(yd_test, pred_svm_diabetes, 'Diabetes - SVM')
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
def plot_roc_curve(models, X_test, y_test, title):
    plt.figure(figsize=(8, 6))

    for name, model in models.items():
        if hasattr(model, "predict_proba"):
            y_proba = model.predict_proba(X_test)[:, 1]
        elif hasattr(model, "decision_function"):
            y_proba = model.decision_function(X_test)
        else:
            continue

        fpr, tpr, _ = roc_curve(y_test, y_proba)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], 'k--')  # Diagonal baseline
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid()
    plt.show()
```

```python
# Models dictionary
heart_models = {
    'Logistic Regression': lr_heart,
    'Random Forest': rf_heart,
    'SVM': svm_heart
}

diabetes_models = {
    'Logistic Regression': lr_diabetes,
    'Random Forest': rf_diabetes,
    'SVM': svm_diabetes
}

# Plot ROC curves
plot_roc_curve(heart_models, Xh_test, yh_test, 'ROC Curve - Heart Disease Models')
plot_roc_curve(diabetes_models, Xd_test, yd_test, 'ROC Curve - Diabetes Models')
```

# Evaluation and comparison of models

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def evaluate_models(models, X_test, y_test):
    results = []
    for name, model in models.items():
        y_pred = model.predict(X_test)

        if hasattr(model, "predict_proba"):
            y_proba = model.predict_proba(X_test)[:, 1]
        elif hasattr(model, "decision_function"):
            y_proba = model.decision_function(X_test)
        else:
            y_proba = None

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred)
        rec = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_proba) if y_proba is not None else "N/A"

        results.append([name, acc, prec, rec, f1, auc])
    
    return results
```

```python
import pandas as pd

# Heart Models
heart_results = evaluate_models(heart_models, Xh_test, yh_test)
df_heart = pd.DataFrame(heart_results, columns=["Model", "Accuracy", "Precision", "Recall", "F1-Score", "AUC"])
print("Heart Disease Model Comparison:\n")
print(df_heart.sort_values("F1-Score", ascending=False))

# Diabetes Models
diabetes_results = evaluate_models(diabetes_models, Xd_test, yd_test)
df_diabetes = pd.DataFrame(diabetes_results, columns=["Model", "Accuracy", "Precision", "Recall", "F1-Score", "AUC"])
print("\nDiabetes Model Comparison:\n")
print(df_diabetes.sort_values("F1-Score", ascending=False))
```
